```python
# 1. Load the "students.csv" dataset using pandas.


import pandas as pd
from scipy import stats


df = pd.read_csv('students.csv')            
df.head()

# Once we've executed the code, the dataset will be loaded into a pandas DataFrame called df.
# then we can use df.head() to display the first few rows of the dataset

```

```python
# 2. Perform the following data-cleaning tasks:
# a. Identify and handle missing values:
# ● Check for missing values in the dataset.
# ● Decide on an appropriate strategy for handling missing values (e.g., dropping
# rows/columns, imputation).
# ● Implement the chosen strategy and update the dataset.

import pandas as pd

missing_values = df.isnull().sum()
print(missing_values)


# To check for missing values in the dataset, we can use the isnull() function followed by the sum() function 
# to get the count of missing values for each column:
```

```python
# 3 b. Remove duplicate records:
# ● Identify any duplicate records in the dataset.
# ● Remove the duplicate records and update the dataset accordingly.

import pandas as pd

# Identify duplicate records
duplicate_rows = df.duplicated()
print("Number of duplicate records:", duplicate_rows.sum())

# Remove duplicate records
df_unique = df[~duplicate_rows].copy()

# Verify if duplicate records have been removed
print("Number of records after removing duplicates:", len(df_unique))
```

```python
# 4. c. Fix data inconsistencies:
# ● Identify any inconsistencies or errors in the dataset (e.g., inconsistent
# capitalization, misspelled values).
# ● Correct the inconsistencies and update the dataset.

import pandas as pd

df = df.rename(columns={'gender': 'Gender'})
df = df.rename(columns={'NationalITy': 'Nationality'})
df = df.rename(columns={'raisedhands': 'RaisedHands'})
df = df.rename(columns={'VisITedResources': 'VisitedResources'})

df['Nationality'] = df['Nationality'].replace('lebanon', 'Lebanon')
df['Nationality'] = df['Nationality'].replace('venzuela', 'Venezuela')
df['Nationality'] = df['Nationality'].replace('Tunis', 'Tunisia')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('KuwaIT', 'Kuwait')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('venzuela', 'Venezuela')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('lebanon', 'Lebanon')
df['PlaceofBirth'] = df['PlaceofBirth'].replace('Tunis', 'Tunisia')
df['StageID'] = df['StageID'].replace('lowerlevel', 'LowerLevel')
df['Relation'] = df['Relation'].replace('Mum', 'Mother')
df.head(10)

# First I have renamed the misspelled columns using rename() function and then correct the spellings using 
# replace()function
```

```python
# 5. d. Convert data types:
# ● Check the data types of columns and ensure they are appropriate for their
# respective data.
# ● Convert any columns with incorrect data types to the correct ones (e.g., numeric
# columns to float/int, dates to datetime).


import pandas as pd

df.dtypes

# .dtypes() function wil show the data types of every column in dataset, but here we havent change or modified
# any datatype of column bcoz they are in correct format 
```

```python
# 6. e. Handle outliers:
# ● Identify any outliers in the dataset (e.g., extremely high or low values).
# ● Decide on an appropriate approach to handle the outliers (e.g., removing outliers,
# capping/extending values).
# ● Implement the chosen approach and update the dataset.

import pandas as pd

select_column1 = 'RaisedHands'
outliers = df[(df[select_column1] < df[select_column1].mean() - 8 * df[select_column1].std()) | (df[select_column1] < df[select_column1].mean() + 8 * df[select_column1].std())]

select_column2 ='VisitedResources'
outliers = df[(df[select_column2] < df[select_column2].mean() - 8 * df[select_column2].std()) | (df[select_column2] < df[select_column2].mean() + 8 * df[select_column2].std())]

select_column3 ='AnnouncementsView'
outliers = df[(df[select_column3] < df[select_column3].mean() - 8 * df[select_column3].std()) | (df[select_column3] < df[select_column3].mean() + 8 * df[select_column3].std())]

select_column4 ='Discussion'
outliers = df[(df[select_column4] < df[select_column4].mean() - 8 * df[select_column4].std()) | (df[select_column4] < df[select_column4].mean() + 8 * df[select_column4].std())]


print('Outliers: \n')
outliers
```
